Como trabajo de laboratorio vamos a entrenar un modelo de regresión para el caso **univariado** $y_n = f(x_n) + \epsilon$. Llevar a cabo el siguiente procedimiento:

1. Descargar la base de datos ```EET-small``` y visualizar la variable a predecir ```oil temperature``` (https://github.com/zhouhaoyi/ETDataset).

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import random
from sklearn.model_selection import train_test_split
````

````python
data = pd.read_csv('ETTh1.csv')
data
````

````python
fechas = data['date'].values
X = np.linspace(0,len(fechas),len(fechas)) #datos
Y = data['OT'].values # etiquetas
Y = Y / np.max(Y)
````

````python
plt.figure(figsize=(10,4))
plt.plot(X,Y)
plt.title('Temperatura del aceite')
plt.ylabel('OT')
plt.xlabel('datos')
plt.grid()
````

2. Dividir la base de datos en 70% entrenamiento y 30% test, utilizando una partición aleatoria.


````python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size =0.7,random_state=42) #Se divide la base de datos 70-30

plt.figure(figsize=(10,4))
plt.plot(X_train, Y_train, 'or')
plt.plot(X_test, Y_test, 'oc')
plt.plot(X, Y, '.k')
plt.title('Temperatura del aceite (OT)')
plt.legend(['datos entrenamiento', 'datos test', 'todos los datos'])
plt.grid()
plt.show()
````

3. Utilice las funciones exponenciales como funciones base, y estime los parámetros del modelo utilizando MAP y MLE para diez valores diferentes de $M$ (número de funciónes base).


````python
def expo_carac(X, K):  #Funciones base
    # Extraer caracteristicas exponenciales
    # X: N x D, vector de entradas
    # K: escalar, grado del polinomio
    # return: N x (K+1), matriz de características Phi

    X = X.flatten()
    N = X.shape[0]

    #inicializar Phi
    Phi = np.ones((N, K+1))
    mu = np.linspace(0,len(X),K)
    sig2 = 6000

    # Calcular matriz Phi
    for k in range(1, K+1):
        Phi[:,k] = np.exp(-(mu[k-1] - X)**2/sig2)
    return Phi

def inferencia_max_vero_carac(Phi, y): #ML
    # máxima verosimilitud usando caracteristicas
    # Phi: N x D, caracteristicas
    # y: N x 1, datos de salida observados
    # return: D x 1, vector de parametros estimados

    kappa = 1e-08 # jitter
    D = Phi.shape[1]

    Pt = Phi.T @ y # Phi^T*y
    PP = Phi.T @ Phi + kappa*np.eye(D) # Phi^T*Phi + kappa*I
    C = sp.linalg.cho_factor(PP)
    theta_ml = sp.linalg.cho_solve(C, Pt) # inv(Phi^T*Phi)*Phi^T*y
    return theta_ml

def inferencia_max_posterior_carac(Phi, y, sigma, alpha): #MAP
    # Phi: N x D, características de entrada
    # y: D x 1, datos de salida
    # sigma: escalar, desviación estandar del ruido
    # alpha: escalar, desviación estandar del prior sobre los parámetros
    # return: D x 1,  estimación MAP theta_map

    D = Phi.shape[1]

    # SOLUTION
    PP = Phi.T @ Phi + (sigma/alpha)**2 * np.eye(D)
    theta_map = sp.linalg.solve(PP, Phi.T @ y)

    return theta_map

````

````python
k = [20,40,60,80,100,120,140,170,200,220] #Valores para M

for i in k:
    alpha = 0.5 # standard deviation of the parameter prior
    sigma = 0.3

    # características 
    phi_train= expo_carac(X_train, i) 
    phi_test = expo_carac(X, i) 

    thetha_mle = inferencia_max_vero_carac(phi_train,Y_train)
    theta_map = inferencia_max_posterior_carac(phi_train,Y_train,sigma, alpha)

    # predicción
    y_pred_expo = phi_test @ thetha_mle
    y_pred_map = phi_test @ theta_map 

    plt.figure(figsize=(10,4))
    plt.plot(X,y_pred_expo, lw=2)
    plt.plot(X,y_pred_map, lw=2)
    plt.plot(X, Y, '.k')
    plt.title(f'Prediición con k = {i}')
    plt.legend(["predicción MaxVero","predicción MaxPos", "datos entrenamiento"])
    plt.grid()
    plt.show()


````

4. Calcule el MSE entre:

 Visualice ambos errores en una misma gráfica ($M$ vs MSE-MAP y $M$ vs. MSE-MLE). De acuerdo a los resultados obtenidos establezca:
  - la predicción de los modelos y los datos de entrenamiento.
  - la predicción de los modelos y los datos de test.


````python
def rmse(y, y_pred):
    # Calculo función de error
    y = y.flatten()
    y_pred = y_pred.flatten()
    N = y.shape[0]
    difer = (y - y_pred)**2
    return np.sqrt((1.0/N) * np.sum(difer))
````

````python
rmse_train_mle = []
rmse_test_mle = []

rmse_train_map = []
rmse_test_map = []

K_vector = []

K_max = 60

if K_max <= 100:
    cons=1
else:
    cons=10

for i in range(1, K_max,cons):
    K = i

    Phi_train = expo_carac(X_train, K)

    theta_mle = inferencia_max_vero_carac(Phi_train, Y_train)
    theta_map = inferencia_max_posterior_carac(Phi_train, Y_train,sigma,alpha)

    #Medicion Error
    y_pred_mle = Phi_train @ theta_mle # predicción
    rmse_train_mle.append(rmse(Y_train, y_pred_mle)) # medir error
    y_pred_map = Phi_train @ theta_map # predicción
    rmse_train_map.append(rmse(Y_train, y_pred_map)) # medir error

    # predicción sobre datos de test
    Phi_test = expo_carac(X_test, K)

    #Medicion test
    y_pred_mle = Phi_test @ theta_mle # predicción
    rmse_test_mle.append(rmse(Y_test, y_pred_mle)) # medir error
    y_pred_map = Phi_test @ theta_map # predicción
    rmse_test_map.append(rmse(Y_test, y_pred_map)) # medir error

    K_vector.append(i)

#Maxima verosimilitud
plt.figure(figsize=(15,8))
plt.subplot(2,1,1)
plt.plot(K_vector, rmse_train_mle, '-', label='Error train')
plt.plot(K_vector, rmse_test_mle, '-', label='Error test')
plt.xticks(K_vector)
plt.title('RMSE MaxVero')
plt.ylabel('RMSE')
plt.xlabel('K')
plt.legend()
plt.tight_layout()
plt.grid()

#Maximo a posteriori
plt.subplot(2,1,2)
plt.plot(K_vector, rmse_train_map, '-', label='Error train')
plt.plot(K_vector, rmse_test_map, '-', label='Error test')
plt.xticks(K_vector)
plt.title('RMSE MaxPo')
plt.ylabel('RMSE')
plt.xlabel('K')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
````

  - ¿Cuál es el número en funciones base $M$ correspondiente al modelo con mejor capacidad de generalización?

  El numero de dunciones base M correspondiente al modelo con mejor capacidad de generalización en este caso es de 80 aproximadamente, ya que es el que toma la mayoria de los datos.

  - ¿El MSE entre las predicciónes y los datos de entrenamiento también es mínimo para ese mismo valor de $M$?, argumente.

  Para el metodo de maxima verosimilitud el minimo no es igual, es muy cercano a 60 y en M=80 el error se dispara considerablemente. Sin embargo para el método de máximo posteriori el minimo esta entre 60 y 80, que tiende a estabilizarse, por lo que para M=80 el error no cambia mucho.

  - ¿Qué método (MAP o MLE) obtiene un mejor desempeño respecto a número de funciones base vs. el error?, argumente.

El método que mejor desempeño obtiene respecto al número de funciones base vs el error es el MAP (Maximo a posteriori), debido a que regulariza un poco mejor el error y trata de estandarizar los datos, por lo que se puede concluir que el MAP es un algoritmo mas robusto frente al sobreentrenamiento. El MLE (MAxima verosimilitud) despues de un numero de funciones base aparece un sobreajuste grave por lo que el error se dispara.

5. Por último, utilice regresión lineal Bayesiana y grafique la media posterior y el intervalo de confianza de la distribución predictiva obtenida (use los mismos datos particionados en el punto 2). Utilice diferentes valores para la varianza del prior sobre los parámetros, describa y explique los cambios observados.

````python
## EDIT THIS FUNCTION

def polyfit(X, y, K, prior_var, noise_var):
    # X: training inputs, size N x D
    # y: training targets, size N x 1
    # K: degree of polynomial we consider
    # prior_var: prior variance of the parameter distribution
    # sigma: noise variance

    jitter = 1e-08 # increases numerical stability

    Phi = expo_carac(X, K) # N x (K+1) feature matrix

    # Compute maximum likelihood estimate
    Pt = Phi.T @ y # Phi*y, size (K+1,1)
    PP = Phi.T @ Phi + jitter*np.eye(K+1) # size (K+1, K+1)
    C = sp.linalg.cho_factor(PP)
    # maximum likelihood estimate
    theta_ml = sp.linalg.cho_solve(C, Pt) # inv(Phi^T*Phi)*Phi^T*y, size (K+1,1)

#     theta_ml = scipy.linalg.solve(PP, Pt) # inv(Phi^T*Phi)*Phi^T*y, size (K+1,1)

    # MAP estimate
    theta_map = sp.linalg.solve(PP + noise_var/prior_var*np.eye(K+1), Pt)

    # parameter posterior
    iSN = (np.eye(K+1)/prior_var + PP/noise_var) # posterior precision
    SN = sp.linalg.pinv(noise_var*np.eye(K+1)/prior_var + PP)*noise_var  # posterior covariance
    mN = sp.linalg.solve(iSN, Pt/noise_var) # posterior mean

    return (theta_ml, theta_map, mN, SN)

def regresion_bayesiana(theta_ml, theta_map, theta_mean, theta_var):
# predictions (ignoring the measurement/observations noise)

    Phi_test = expo_carac(X,k) # N x (K+1)

    # maximum likelihood predictions (just the mean)
    m_mle_test = Phi_test @ theta_ml

    # MAP predictions (just the mean)
    m_map_test = Phi_test @ theta_map

    # predictive distribution (Bayesian linear regression)
    # mean prediction
    mean_blr = Phi_test @ theta_mean
    # variance prediction
    cov_blr =  Phi_test @ theta_var @ Phi_test.T

    # plot the posterior
    plt.figure(figsize=(15,6))
    #plt.plot(Xtest, m_mle_test)
    plt.plot(X, m_map_test, 'C0')
    var_blr = np.diag(cov_blr)

    #conf_bound1 = np.sqrt(var_blr).flatten()
    conf_bound2 = 2.0*np.sqrt(var_blr).flatten()
    #conf_bound3 = 2.0*np.sqrt(var_blr + sigma).flatten()

    #plt.fill_between(Xtest.flatten(), mean_blr.flatten() + conf_bound1, mean_blr.flatten() - conf_bound1, alpha = 0.1, color="k")
    plt.fill_between(X.flatten(), mean_blr.flatten() + conf_bound2, mean_blr.flatten() - conf_bound2, alpha = 0.1, color="C0")
    #plt.fill_between(Xtest.flatten(), mean_blr.flatten() + conf_bound3, mean_blr.flatten() - conf_bound3, alpha = 0.1, color="k")


    plt.legend(["media posterior", "varianza posterior"])
    plt.xlabel('$x$');
    plt.ylabel('$y$');
    plt.grid()
````

````python
k = 20
alpa = 0,25
sigma = 0.5
theta_ml, theta_map, theta_mean, theta_var = polyfit(X_train, Y_train,k, alpha, sigma)
regresion_bayesiana(theta_ml, theta_map, theta_mean, theta_var)
````

````python
k = 20
alpa = 0,5
sigma = 0.8
theta_ml, theta_map, theta_mean, theta_var = polyfit(X_train, Y_train,k, alpha, sigma)
regresion_bayesiana(theta_ml, theta_map, theta_mean, theta_var)
````

````python
k = 20
alpa = 1
sigma = 3
theta_ml, theta_map, theta_mean, theta_var = polyfit(X_train, Y_train,k, alpha, sigma)
regresion_bayesiana(theta_ml, theta_map, theta_mean, theta_var)
````

Para la regresiónn bayesiana, se puede apreciar que variando la varianza del prior sobre los parámetros, se observan cambios significativos en los intervalos de confianza y en la estabilidad de las predicciones. Valores más bajos de varianza tienden a restringir el modelo, mientras que valores más altos permiten una mayor flexibilidad, aunque con el riesgo de sobreajuste. Ademas la elección de una varianza adecuada es crucial para equilibrar la capacidad predictiva y la robustez del modelo.

# Conclusiones


- En general, un valor de M demasiado bajo puede resultar en un modelo subajustado (underfitting), mientras que un valor de M demasiado alto puede llevar a un sobreajuste (overfitting).
- El MSE en el conjunto de entrenamiento tiende a disminuir con el aumento de M, ya que el modelo tiene más capacidad para ajustarse a los datos de entrenamiento.
Sin embargo, el MSE en el conjunto de test primero disminuye y luego puede aumentar debido al sobreajuste. El valor óptimo de M para la mejor generalización no necesariamente corresponde al mínimo MSE en el conjunto de entrenamiento, sino al balance óptimo donde el modelo no está ni subajustado ni sobreajustado.
- El método MAP generalmente incorpora un prior que puede regularizar el modelo, evitando el sobreajuste y mejorando la generalización. 
En los resultados, el MAP puede mostrar un mejor desempeño respecto al MLE, especialmente para valores de M altos, ya que la regularización ayuda a controlar la complejidad del modelo.
- Para la regresion lineal Bayesiana se puede concluir que con un prior de alta varianza, el modelo puede ajustarse más a los datos de entrenamiento, pero corre el riesgo de sobreajustarse. Con un prior de baja varianza, el modelo es más robusto y puede generalizar mejor, pero con el riesgo de subajustarse si la regularización es demasiado fuerte.
