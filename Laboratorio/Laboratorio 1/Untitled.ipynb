```python
#librerias a utilizar
import numpy as np
from matplotlib.pylab import *
#from scipy import *
import scipy.signal as signal
get_ipython().run_line_magic('matplotlib', 'inline')
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write #funciones para leer y escribir audio
from IPython.display import Audio # para escuchar la senal
import scipy.signal as sp
from numpy import pad


# Generar una seal de ejemplo (senoidal)
fs = 500  # Frecuencia de muestreo
t = np.arange(0, 1.0, 1.0/fs)
f = 5  # Frecuencia de la seal
signal = np.sin(2 * np.pi * f * t)

# Aplicar la ventana de Hamming
window = hamming(len(signal))
windowed_signal = signal * window

# Graficar la seal original, la ventana de Hamming y la seal enventanada
plt.figure(figsize=(12, 8))

# Graficar la seal original
plt.subplot(3, 1, 1)
plt.plot(t, signal, label='Seal Original')
plt.title('Seal Original')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)

# Graficar la ventana de Hamming
plt.subplot(3, 1, 2)
plt.plot(t, window, label='Ventana de Hamming', color='orange')
plt.title('Ventana de Hamming')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)

# Graficar la seal enventanada
plt.subplot(3, 1, 3)
plt.plot(t, windowed_signal, label='Seal con Ventana de Hamming', color='green')
plt.title('Seal con Ventana de Hamming')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)

plt.tight_layout()
plt.show()
```
