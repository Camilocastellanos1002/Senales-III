### 3. Enventanado

#### A continuacion se desarrolla una funcion que realiza la segmentacion por enventanado de las se√±ales 

```python

def segmento_senal(senal, fs, L, P, tipo_ventana='rectangular'):
    #senal -> Se√±al de entrada
    #fs -> Frecuencia de muestreo
    #L -> Longitud de la ventana
    #P -> Traslape entre ventanas
    #Tipo ventana -> rectangular, hann, hamming, blackman, bartlett

    # Inicializar la lista de segmentos
    segmentos = []
    
    # Validar la longitud de la ventana y el traslape
    if L > len(senal) or P >= L:
        raise ValueError("Longitud de ventana o traslape inv√°lidos.")

    # Seleccion de enventanado
    if tipo_ventana == 'rectangular':
        ventana = np.ones(L)
    elif tipo_ventana == 'hann':
        ventana = np.hanning(L)
    elif tipo_ventana == 'hamming':
        ventana = np.hamming(L)
    elif tipo_ventana == 'blackman':
        ventana = np.blackman(L)
    elif tipo_ventana == 'bartlett':
        ventana = np.bartlett(L)
    else:
        raise ValueError("Tipo de ventana no soportado.")

    # imprimir segmentos
    for i in range(0, len(senal) - L + 1, P):
        segmento = senal[i:i+L] * ventana
        segmentos.append(segmento)

    return segmentos
```

### 4. Reconstruccion de la se√±al por medio del enventanado

#### Se crea la funcion de construccion de la se√±ales

```python
def reconstruir_senal(segmentos, P):
    #segmentos -> segmentos de la se√±al a reconstruir
    #P -> tama√±o de traslape entre segmentos
    
    # Asegurarse que 'segmentos' sea un array de NumPy
    segmentos = np.array(segmentos)
    
    # longitud de un segmento
    L = segmentos.shape[1]
    
    # Tama√±o del arrays, iniciandolo con valores en cero
    senal_reconstruida = np.zeros(L + (segmentos.shape[0] - 1) * P)
    
    # suma de segmentos
    for i in range(segmentos.shape[0]):
        senal_reconstruida[i*P:i*P+L] += segmentos[i]
    
    return senal_reconstruida
```

#### Ahora procedemos a utilizar la funcion para reconstruir cada uno de los segmentos de las diferentes se√±ales

* #### Se√±al #1

```python

# Reconstruir la se√±al
senal_reconstruida = reconstruir_senal(segmentos, P)

# Visualizar la se√±al original y la reconstruida
plt.figure(figsize=(10, 6))
plt.plot(segmento_guardado, label='Se√±al Original')
plt.plot(senal_reconstruida, label='Se√±al Reconstruida')
plt.legend()
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.title('Comparaci√≥n entre Se√±al Original y Reconstruida')
plt.grid(True)
plt.show()
```

```python
1. Efecto de la longitud de la ventana:

Resoluci√≥n en tiempo y frecuencia: Ventanas m√°s cortas ofrecen mejor resoluci√≥n temporal pero peor resoluci√≥n en frecuencia, y viceversa. Esto implica un compromiso entre la capacidad de localizar eventos transitorios y la precisi√≥n en la estimaci√≥n de frecuencias.
Distorsi√≥n de fase: Algunas ventanas introducen mayor distorsi√≥n de fase que otras, lo que puede afectar la percepci√≥n de la se√±al reconstruida, especialmente en aplicaciones de audio.
2. Efecto del traslape:

Suavidad de la reconstrucci√≥n: Un mayor traslape reduce los efectos de borde y produce una reconstrucci√≥n m√°s suave, pero aumenta la redundancia en los c√°lculos.
Complejidad computacional: Un mayor traslape implica un mayor n√∫mero de operaciones y, por lo tanto, un mayor costo computacional.
3. Comparaci√≥n de diferentes m√©tricas:

RMSE (Root Mean Square Error): Mide la diferencia cuadr√°tica media entre la se√±al original y la reconstruida. Es sensible a errores grandes.
SNR (Signal-to-Noise Ratio): Mide la relaci√≥n entre la potencia de la se√±al y la potencia del ruido. Es √∫til para evaluar la calidad de la reconstrucci√≥n en presencia de ruido.
PESQ (Perceptual Evaluation of Speech Quality): Es una m√©trica subjetiva que eval√∫a la calidad percibida de la se√±al de voz.
4. An√°lisis en el dominio de la frecuencia:

Espectrogramas: Utiliza la transformada de Fourier de corta duraci√≥n (STFT) para visualizar la evoluci√≥n de la frecuencia en el tiempo.
Distorsi√≥n arm√≥nica total (THD): Mide la distorsi√≥n arm√≥nica introducida por el proceso de segmentaci√≥n y reconstrucci√≥n.
5. Aplicaciones espec√≠ficas:

Compresi√≥n de audio: La segmentaci√≥n y reconstrucci√≥n son fundamentales en algoritmos de compresi√≥n como MP3 y AAC.
An√°lisis de vibraciones: Se utiliza para analizar se√±ales vibratorias en maquinaria y detectar fallas.
Procesamiento de se√±ales biom√©dicas: Se aplica en el an√°lisis de se√±ales EEG, ECG, etc.
Comunicaciones: Se utiliza en sistemas de comunicaciones para modular y demodular se√±ales.
Conclusiones generales:

No existe una ventana √≥ptima para todas las se√±ales y aplicaciones: La elecci√≥n de la ventana depende de las caracter√≠sticas de la se√±al y de los requisitos de la aplicaci√≥n.
El traslape es un par√°metro importante: Influye en la suavidad de la reconstrucci√≥n y en la eficiencia computacional.
La evaluaci√≥n de la calidad de la reconstrucci√≥n debe ser multidimensional: No basta con utilizar una sola m√©trica.
La segmentaci√≥n y reconstrucci√≥n son herramientas fundamentales en el procesamiento de se√±ales: Permiten analizar y modificar se√±ales de manera flexible.
```
