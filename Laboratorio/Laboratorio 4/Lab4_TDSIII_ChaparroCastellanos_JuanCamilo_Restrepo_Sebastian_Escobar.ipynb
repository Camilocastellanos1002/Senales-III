```python
# Cargar el archivo MP3 y especificar solo los primeros 10 segundos
file_audio = 'voz.mp3'
x, fs = librosa.load(file_audio, sr=None, duration=10)

# Reproducir el audio cargado
Audio(x, rate=fs)
```

## 2) Estandarizar la señal de audio

```python
# Estandarizar la señal: media = 0, desviación estándar = 1
mean = np.mean(x) 
std = np.std(x)
x_estandarizado = (x - mean) / std
```

```python
print("Media: ",mean)
print("Desviacion estandar: ",std)

# Reproducir el audio cargado
Audio(x_estandarizado, rate=fs)
```
