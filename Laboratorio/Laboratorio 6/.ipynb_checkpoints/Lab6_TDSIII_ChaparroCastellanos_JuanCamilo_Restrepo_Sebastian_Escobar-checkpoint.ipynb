#### Luego de la lectura del datasset antes de proceder a operar y analizar normalizamos los datos y estandarizarlos

```python
#Funcion para normalizar
def normalize_min_max(data):
    min_val = np.min(data)
    max_val = np.max(data)
    normalized_data = (data - min_val) / (max_val - min_val)
    return normalized_data

#Funcion para estandarizar
def estandarizar(datos):
    media = np.mean(datos)
    sigma = np.std(datos)
    datos_estandarizados = (datos - media) / sigma
    return datos_estandarizados
```

```python
#solo lectura de datos numericos
data_numeric = data.select_dtypes(include=[np.number])

# normalizamos  y estandarizamos
data_normalized = data_numeric.apply(normalize_min_max, axis=0)
data_standardized = data_numeric.apply(estandarizar, axis=0)

# resultados
print("Datos Normalizados:\n", data_normalized.head())
print("Datos Estandarizados:\n", data_standardized.head())
```

## 2. Division de base de datos

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Partición aleatoria
train_random, test_random = train_test_split(data, test_size=0.3, random_state=42)

# Partición secuencial
n_train = int(0.7 * len(data))
train_sequential = data.iloc[:n_train]
test_sequential = data.iloc[n_train:]

# Graficar las particiones
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Graficar partición aleatoria
ax[0].plot(train_random.index, np.ones(len(train_random)), 'o', color='blue', label='Entrenamiento')
ax[0].plot(test_random.index, np.ones(len(test_random)), 'o', color='orange', label='Prueba')
ax[0].set_title("Partición Aleatoria")
ax[0].legend()
ax[0].set_yticks([])

# Graficar partición secuencial
ax[1].plot(train_sequential.index, np.ones(len(train_sequential)), 'o', color='blue', label='Entrenamiento')
ax[1].plot(test_sequential.index, np.ones(len(test_sequential)), 'o', color='orange', label='Prueba')
ax[1].set_title("Partición Secuencial")
ax[1].legend()
ax[1].set_yticks([])

plt.show()
```
